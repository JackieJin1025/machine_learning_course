```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
path = os.getcwd() + '/ex1data1.txt'
data = pd.read_csv(path, header=None, names=['Population', 'Profit'])
data.head()
```

```python
data.describe()
```

```python
data.plot(kind='scatter', x='Population', y='Profit', figsize=(12,8))
```

```python
data.insert(0, 'Ones', 1)
```

```python
cols = data.shape[1]
X = data.iloc[:,0:cols-1].values
y = data.iloc[:,cols-1:cols].values
y = np.ravel(y) 
```

```python
def computeCost(X, y, theta):
    inner = np.power(X.dot(theta) - y, 2)    
    return np.sum(inner) / (2 * len(X))

def gradientDescent(X, y, theta0, alpha, iters):
    theta = theta0.copy()
    parameters =  len(theta)
    cost = np.zeros(iters)

    for i in range(iters):
        error = X.dot(theta) - y
        for j in range(parameters):
            term = error.dot(np.ravel(X[:, j]))                        
            theta[j] = theta[j] - (alpha / len(X) * term )
        
        cost[i] = computeCost(X, y, theta)

    return theta, cost 
```

```python
iters = 1000
alpha = 0.01
theta0 = np.array([0.0,0.0])
theta, cost = gradientDescent(X, y, theta0, alpha, iters)
print(computeCost(X, y, theta0))
print(computeCost(X, y, theta))
```

```python
x = np.linspace(data.Population.min(), data.Population.max(), 100)
```

```python
f = theta[0] + theta[1] * x
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, f, 'r', label='Prediction')
ax.scatter(data.Population, data.Profit, label='Traning Data')
ax.legend(loc=2)
ax.set_xlabel('Population')
ax.set_ylabel('Profit')
ax.set_title('Predicted Profit vs. Population Size')
```

```python
fig, ax = plt.subplots(figsize=(16,12))

ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title("Error vs Training Epoch with learning rate")

alphas = [ 0.01, 0.001, 0.0001] # 0.1 diverge 
for i in range(len(alphas)):
    alpha = alphas[i]
    theta, cost = gradientDescent(X, y, theta0, alpha, iters)
    ax.plot(np.arange(iters), cost, label="learning rate %s" % alpha  )
ax.legend()    
```

# Linear regression with multiple variables

```python
path = os.getcwd() + '/ex1data2.txt'
data2 = pd.read_csv(path, header=None, names=['Size', 'Bedrooms', 'Price'])
m = data2.mean()
s = data2.std()

data2 = (data2 - m)/s 

data2.insert(0, 'Ones', 1)
data2.head()
```

```python
cols = data2.shape[1]
X2 = data2.iloc[:, 0:cols-1].values
y2 = np.ravel(data2.iloc[:, cols-1:].values) 
```

```python
y2
```

```python
fig, ax = plt.subplots(figsize=(16,12))

ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title("Error vs Training Epoch with learning rate")

alphas = [ 0.01, 0.001, 0.0001] # 0.1 diverge 
theta0_2 = [ 0.0, 0.0, 0.0]
iters2 = 10000
for i in range(len(alphas)):
    alpha = alphas[i]
    theta, cost = gradientDescent(X2, y2, theta0_2, alpha, iters2)
    ax.plot(np.arange(iters2), cost, label="learning rate %s" % alpha  )
ax.legend() 
```

```python
theta, cost = gradientDescent(X2, y2, theta0_2, 0.01, iters2)
print(theta)
```
